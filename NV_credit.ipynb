```
data=pd.read_csv("credit_edit.csv", header=0)
```

```
feature_cols = ['Income', 'Limit', 'Rating', 'Cards',
       'Age', 'Education', 'Gender', 'Student', 'Married', 'Ethnicity',
       'Balance', 'Defaultee']
X = data[feature_cols] # Features
y = data.Defaultee # Target variable


#X = data.loc[:, data.columns != 'Defaultee']
#y = data.loc[:, data.columns == 'Defaultee']
X
y
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=0)
columns = X_train.columns
columns
```

```
#basically this is our fit function :)
# statistics for the train set
print(X_train)
X_train.columns.get_level_values(0)
print(X_train.groupby(y_train))

means = X_train.groupby(y_train, as_index=False).apply(np.mean)
stds = X_train.groupby(y_train).apply(np.std)
print(means, stds)

#class prior probabilities
probs = X_train.groupby(y_train).apply(lambda x: len(x)) / X_train.shape[0]
print(probs)
```

```
y_pred = []

#for each element in the valdation set
for elem in range(X_val.shape[0]):
  p={}
  # for each possible class
  for c1 in np.unique(y_train):
  
    #take the prior probability of the given class
    p[c1] = probs.iloc[c1]  
    # for each column in the data
    for index, param in enumerate(X_val.iloc[elem]):
      print(index, param)
      p[c1] *= norm.pdf(param, means.iloc[c1, index], stds.iloc[c1, index])
    print(p)
  y_pred.append(pd.Series(p).values.argmax())
  print(y_pred)
```

```
from sklearn.metrics import accuracy_score
accuracy_score(y_val, y_pred)
```
